```python
df.shape
```

```python
# Using If Statements

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range '
    else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
# if statements with the loc function
```

```python
df.loc[df['prices'] >15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Repeating process on entire dataframe

df_ords_prods_merged.loc[df_ords_prods_merged['prices'] >15, 'price_range_loc'] = 'High-range product'
```

```python
# Create new column in dataframe
df_ords_prods_merged['order_days_of_week'].value_counts(dropna = False)
```

```python
# Set the loop

result = []

for value in df_ords_prods_merged["order_days_of_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
            result.append("Least Busy")
    else:
                result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merged["busiest_day"] = result
```

```python
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

```python
# Step 2 - Creating new loop for busiest days

result_v2 = []

for value in df_ords_prods_merged["order_days_of_week"]:
    if value == 0 or value == 1:
        result_v2.append("Busiest days")
    elif value == 4 or value == 3:
        result_v2.append("Least busy days")
    else:
        result_v2.append("Regularly busy days")
```

```python
df_ords_prods_merged['Busiest_days'] = result_v2
```

```python
# Step 3 Check the values of this column

df_ords_prods_merged['Busiest_days'].value_counts(dropna = False)
```

```python
# Step 4 Creating a new for loop column Busiest Hours first we have to check the busiest hours of the day

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
result_v3 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_v3.append("Most orders")
    elif value in [3, 4, 2, 5, 1, 0, 6, 23]:
        result_v3.append("Fewest orders")
    else:
        result_v3.append("Average orders")
```

```python
df_ords_prods_merged['Busiest_period_of_day'] = result_v3
```
