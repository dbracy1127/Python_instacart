```python
# To recap, you should always use the groupby() function as part of a series of steps, namely, the following:

# 1. Split the data into groups based on some criteria.
# 2. Apply a function to each group separately.
# 3. Combine the results into a dataframe or alternative data structure or create a new column in the current dataframe.

# So far, you’ve only completed the first step—splitting the data into groups (with the group being the “product_name” column)
```

```python
#  If you were to calculate the mean of the “order_number” column grouped by the “department_id” column,
# you could quickly compare the average number of orders across each Instacart department.
# This will involve two of the steps in the three-step process introduced above:

# 1. Split the data into groups based on “department_id.”
# 2. Apply the agg() function to each group to obtain the mean values for the “order_number” column.
```

```python
df.groupby('department_id').agg({'order_number':['mean']})
```

```python
# An additional way of findig mean

df.groupby('department_id')['order_number'].mean()
```

```python
# Now, let’s map this task onto the three-step process introduced earlier:

# 1. Split the data into groups based on the “user_id” column.
# 2. Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# 3. Create a new column, “max_order,” into which you’ll place the results of your aggregation.
```

```python
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merged.head(20)
```

```python
# Deriving columns with loc
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'Loyalty_flag'] = 'Loyal Customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'Loyalty_flag'] = 'Regular Customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'Loyalty_flag'] = 'New Customer'
```

```python
df_ords_prods_merged['median_days'] = df_ords_prods_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['median_days'] <= 20) & (df_ords_prods_merged['median_days'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merged['frequency_flag'].value_counts(dropna = False)
```

```python
# Changing the null values in "frequency_flag" column to "Non-frequent customer"
df_ords_prods_merged['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```
